```python
import re
import unicodedata
import pandas as pd

def normalize_text(x: str) -> str:
    if pd.isna(x):
        return ""
    x = str(x)
    x = unicodedata.normalize("NFKC", x)          # 全半角など
    x = re.sub(r"\s+", " ", x).strip()            # 連続空白
    x = re.sub(r"[\u2026…]+", "", x)              # 省略記号
    x = re.sub(r"\（[^）]*補完情報[^）]*\）", "", x) # （補完情報）系を削る
    x = re.sub(r"\[[^\]]*中略[^\]]*\]", "", x)     # [中略]系
    return x.strip()

# df は既にある前提
for c in ["title", "fail_factor", "impact"]:
    df[c] = df[c].map(normalize_text)

df["confidence"] = pd.to_numeric(df["confidence"], errors="coerce")
df = df.dropna(subset=["fail_factor", "impact", "confidence"])
```

```python
df = df[df["confidence"] >= 0.8].copy()  # 最初は厳しめ推奨

df["text_for_embed"] = (
    "原因: " + df["fail_factor"] + " / 結果: " + df["impact"]
)

# 長すぎるものを軽くカット（モデル次第で調整）
df["text_len"] = df["text_for_embed"].str.len()
```

```python
import re

def normalize(text: str) -> str:
    if text is None:
        return ""
    t = text
    # 年号・数字の正規化（型で寄るのを抑える）
    t = re.sub(r"\b(19|20)\d{2}年\b", "<YEAR>年", t)
    t = re.sub(r"\b\d+(\.\d+)?\b", "<NUM>", t)
    # 余計な空白整理
    t = re.sub(r"\s+", " ", t).strip()
    return t

df["text_for_embed"] = df["text_for_embed"].map(normalize)
```

```python
import numpy as np
import pandas as pd

project_id = "file"
cluster_col = "cluster_id"
sector_col  = "分野"          # あなたの列名に合わせて
region_col  = "region_detail" # あなたの列名に合わせて

# 案件メタ（案件ごとに分野・地域は1つという前提）
proj_meta = (
    df.dropna(subset=[project_id])
      .drop_duplicates(project_id)[[project_id, sector_col, region_col]]
      .set_index(project_id)
)

# presence（案件内でそのクラスタが出たか 0/1）
presence = (
    df.assign(pres=1)
      .pivot_table(index=project_id, columns=cluster_col, values="pres",
                   aggfunc="max", fill_value=0)
)

# 分析用テーブル（案件単位）
P = proj_meta.join(presence, how="inner")
```

```python
from scipy.stats import fisher_exact

def bh_fdr(pvals: np.ndarray) -> np.ndarray:
    """Benjamini-Hochberg FDR"""
    p = np.asarray(pvals, dtype=float)
    n = len(p)
    order = np.argsort(p)
    ranked = p[order]
    q = ranked * n / (np.arange(n) + 1)
    q = np.minimum.accumulate(q[::-1])[::-1]
    out = np.empty_like(q)
    out[order] = np.clip(q, 0, 1)
    return out

def group_cluster_lift_table(P: pd.DataFrame, group_col: str, clusters: list,
                             min_n_projects: int = 15) -> pd.DataFrame:
    # base rate（全体）
    base = P[clusters].mean()

    rows = []
    for g, sub in P.groupby(group_col):
        n = len(sub)
        if n < min_n_projects:
            continue

        rate = sub[clusters].mean()
        lift = (rate / base).replace([np.inf, -np.inf], np.nan)

        # 参考：2x2 Fisher のp値（多重検定するので後でFDR）
        pvals = []
        for c in clusters:
            a = int(sub[c].sum())        # group内で clusterあり
            b = int(n - a)               # group内で clusterなし
            A = int(P[c].sum())          # 全体で clusterあり
            N = int(len(P))
            c1 = int(A - a)              # group外で clusterあり
            d1 = int((N - A) - b)        # group外で clusterなし
            # [[a,b],[c1,d1]]
            _, pval = fisher_exact([[a,b],[c1,d1]], alternative="two-sided")
            pvals.append(pval)

        qvals = bh_fdr(np.array(pvals))

        for i, c in enumerate(clusters):
            rows.append({
                "group": g,
                "n_projects": n,
                "cluster": c,
                "rate": float(rate[c]),
                "base": float(base[c]),
                "lift": float(lift[c]),
                "p_fisher": float(pvals[i]),
                "q_fdr": float(qvals[i]),
            })

    out = pd.DataFrame(rows)
    return out

# clusters（列名が int のままなら、そのまま）
clusters = list(presence.columns)

sector_tbl = group_cluster_lift_table(P, sector_col, clusters, min_n_projects=15)
region_tbl = group_cluster_lift_table(P, region_col, clusters, min_n_projects=15)

# 各分野の上位（lift順、かつ最低限 rate もある程度、FDRも参考）
top_sector = (
    sector_tbl.sort_values(["group", "lift"], ascending=[True, False])
              .groupby("group").head(10)
)

top_region = (
    region_tbl.sort_values(["group", "lift"], ascending=[True, False])
              .groupby("group").head(10)
)
```

```python
def from_aggregated_table(tbl, region_col="group", cluster_col="cluster",
                          value_col="lift", log2=True, min_count=5,
                          count_col=None):
    mat = tbl.pivot(index=region_col, columns=cluster_col, values=value_col)

    if log2:
        mat = np.log2(mat.replace(0, np.nan))

    counts = None
    if count_col is not None:
        counts = tbl.pivot(index=region_col, columns=cluster_col, values=count_col)
        mat = mat.mask(counts < min_count)

    return mat, counts
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import japanize_matplotlib

from scipy.cluster.hierarchy import linkage, leaves_list
from scipy.spatial.distance import pdist

def make_region_cluster_mats(df, region_col="region_detail", cluster_col="cluster", min_count=5):
    # counts: 地域×クラスタの件数
    counts = pd.crosstab(df[region_col], df[cluster_col])

    # 地域ごとの総件数
    n_region = counts.sum(axis=1)

    # 比率（地域内）
    rate = counts.div(n_region, axis=0)

    # 全体のクラスタ基準率
    base = counts.sum(axis=0) / counts.values.sum()

    # lift と log2 lift
    lift = rate.div(base, axis=1)
    log2_lift = np.log2(lift.replace(0, np.nan))

    # 疎なセルをマスク（表示しない）
    mask = counts < min_count
    log2_lift_masked = log2_lift.mask(mask)

    return counts, rate, base, lift, log2_lift_masked, mask

def cluster_order(mat, axis=0, method="average", metric="euclidean"):
    """
    mat: DataFrame
    axis=0: row clustering, axis=1: col clustering
    """
    arr = mat.to_numpy()
    if axis == 0:
        X = np.nan_to_num(arr, nan=0.0)
    else:
        X = np.nan_to_num(arr.T, nan=0.0)

    if X.shape[0] <= 1:
        return list(range(X.shape[0]))

    Z = linkage(pdist(X, metric=metric), method=method)
    return leaves_list(Z).tolist()

def plot_heatmap(mat, title, xlabel, ylabel, vcenter=0.0, cmap="coolwarm",
                 annotate_counts=None, figsize=(12,6)):
    """
    mat: DataFrame (rows=regions, cols=clusters)
    annotate_counts: 同shapeのDataFrameを渡すとセルに件数表示
    """
    data = mat.to_numpy()
    fig, ax = plt.subplots(figsize=figsize)

    # 発散系にしたいので、範囲は中心対称に寄せる（log2 lift向け）
    vmax = np.nanmax(np.abs(data - vcenter))
    vmin, vmax = vcenter - vmax, vcenter + vmax

    im = ax.imshow(data, aspect="auto", vmin=vmin, vmax=vmax, cmap=cmap)

    ax.set_title(title)
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)

    ax.set_xticks(np.arange(mat.shape[1]))
    ax.set_xticklabels(mat.columns, rotation=90)
    ax.set_yticks(np.arange(mat.shape[0]))
    ax.set_yticklabels(mat.index)

    fig.colorbar(im, ax=ax, fraction=0.02, pad=0.02)

    # 件数注記（必要なら）
    if annotate_counts is not None:
        cnt = annotate_counts.to_numpy()
        for i in range(mat.shape[0]):
            for j in range(mat.shape[1]):
                if not np.isnan(data[i, j]):
                    ax.text(j, i, str(int(cnt[i, j])), ha="center", va="center", fontsize=8)

    plt.tight_layout()
    plt.show()

# ==== 実行例 ====
counts, rate, base, lift, log2_lift, mask = make_region_cluster_mats(df, min_count=5)

# 行（地域）はクラスタリング順のまま
row_order = cluster_order(log2_lift.fillna(0.0), axis=0)

# 列（クラスタ）は昇順に固定（数値としてソート）
# clusterがintならそのまま、strなら数値変換してソート
try:
    col_order = np.argsort(log2_lift.columns.astype(int)).tolist()
except Exception:
    # 数値化できないケースのフォールバック（辞書順）
    col_order = np.argsort(log2_lift.columns.astype(str)).tolist()

log2_lift_ord = log2_lift.iloc[row_order, :].iloc[:, col_order]
counts_ord    = counts.iloc[row_order, :].iloc[:, col_order]
rate_ord      = rate.iloc[row_order, :].iloc[:, col_order]

plot_heatmap(
    log2_lift_ord, "Region × Cluster (log2 lift, masked)", "Cluster", "Region",
    annotate_counts=counts_ord, figsize=(14,7)
)

plot_heatmap(
    rate_ord, "Region × Cluster (rate)", "Cluster", "Region",
    vcenter=0.0, cmap="viridis", annotate_counts=None, figsize=(14,7)
)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import japanize_matplotlib

def plot_value_by_cluster_heatmap(
    df, cluster_col, meta_col,
    normalize="cluster",   # "cluster"=各cluster内で割合(列和=1), "value"=各カテゴリ値内で割合(行和=1), None=件数
    annotate=True,
    min_count=0,           # 件数がこれ未満のセルはマスク
    figsize=(14, 6),
    cmap="viridis",
):
    d = df[[cluster_col, meta_col]].copy()
    d[meta_col] = d[meta_col].astype("string").fillna("NA")

    # 件数：cluster×value
    ct = pd.crosstab(d[cluster_col], d[meta_col])

    # clusterを昇順に
    ct = ct.sort_index(axis=0)

    # 正規化
    if normalize == "cluster":
        mat = ct.div(ct.sum(axis=1), axis=0)  # cluster内比率
        title_suffix = "（cluster内割合）"
        ann = ct if annotate else None
    elif normalize == "value":
        mat = ct.div(ct.sum(axis=0), axis=1)  # value内比率
        title_suffix = "（カテゴリ値内割合）"
        ann = ct if annotate else None
    else:
        mat = ct
        title_suffix = "（件数）"
        ann = None

    # 縦=カテゴリ値, 横=cluster にするため転置
    mat_t = mat.T
    ct_t = ct.T

    # マスク
    if min_count > 0:
        mat_t = mat_t.mask(ct_t < min_count)

    data = mat_t.to_numpy()

    fig, ax = plt.subplots(figsize=figsize)
    im = ax.imshow(data, aspect="auto", cmap=cmap)

    ax.set_title(f"{meta_col} × Cluster {title_suffix}")
    ax.set_xlabel("Cluster")
    ax.set_ylabel(meta_col)

    ax.set_xticks(np.arange(mat_t.shape[1]))
    ax.set_xticklabels(mat_t.columns, rotation=90)
    ax.set_yticks(np.arange(mat_t.shape[0]))
    ax.set_yticklabels(mat_t.index)

    plt.colorbar(im, ax=ax, fraction=0.02, pad=0.02)

    # 注記（件数）
    if ann is not None:
        ann_t = ann.T.to_numpy()
        for i in range(mat_t.shape[0]):
            for j in range(mat_t.shape[1]):
                if not np.isnan(data[i, j]):
                    ax.text(j, i, str(int(ann_t[i, j])), ha="center", va="center", fontsize=8)

    plt.tight_layout()
    plt.show()


# 使い方（例）
cluster_col = "cluster"
plot_value_by_cluster_heatmap(df, cluster_col, "phase", normalize="cluster", annotate=True, min_count=0, figsize=(14,4))
plot_value_by_cluster_heatmap(df, cluster_col, "actor", normalize="cluster", annotate=True, min_count=0, figsize=(14,5))
plot_value_by_cluster_heatmap(df, cluster_col, "controllability", normalize="cluster", annotate=True, min_count=0, figsize=(14,5))
plot_value_by_cluster_heatmap(df, cluster_col, "is_overrun_related", normalize="cluster", annotate=True, min_count=0, figsize=(14,4))
```
