## 機械学習用にプロジェクト開始年起点の特徴量に付け替える

```python
import pandas as pd

CSV_PATH   = "../df_check_99.csv"   # あなたのCSVパスに変更可
df = pd.read_csv(CSV_PATH)
df
```

```python
# 開始時の値結合
df_freedom = pd.read_csv('../out_csv/freedomrate.csv')
df_freedom = df_freedom.rename(columns={'Year':'project_start_year'})
df=df.drop('freedom_house_score', axis=1)
df=pd.merge(df, df_freedom, on=['project_start_year','alpha3'], how='left')
len(df)
```

```python
# 開始時の値結合
df_pop = pd.read_csv('../out_csv/population.csv')
df_pop = df_pop.rename(columns={'Year':'project_start_year'})

df=df.drop('population', axis=1)
df=pd.merge(df, df_pop, on=['project_start_year','alpha3'], how='left')
len(df)
```

```python
# 開始時の値結合
df_gdp = pd.read_csv('../out_csv/gdp.csv')
df_gdp = df_gdp.rename(columns={'Year':'project_start_year'})

df=df.drop(['gdp_growth','join_flg','Unnamed: 0'], axis=1)
df=pd.merge(df, df_gdp, on=['project_start_year','alpha3'],how='left')
len(df)
```

```python
# 開始時の値結合
wgi = [
    'control_of_corruption', 
    'government_effectiveness', 
    'political_stability', 
    'regulatory_quality', 
    'rule_of_law', 
    'voice_and_accountability',     
]

rename_dict={
    'Control of Corruption': 'control_of_corruption',
    'Government Effectiveness': 'government_effectiveness',
    'Political Stability': 'political_stability',
    'Regulatory Quality': 'regulatory_quality',
    'Rule of Law': 'rule_of_law',
    'Voice and Accountability': 'voice_and_accountability',
    'country': 'alpha3',
    'year': 'project_start_year',
}

df_wgi = pd.read_csv('../out_csv/wgi.csv')
df_wgi = df_wgi.rename(columns=rename_dict)
df=df.drop(columns=wgi, axis=1)
df=df.drop(['join_flg','Unnamed: 0'], axis=1)

df=pd.merge(df, df_wgi, on=['project_start_year','alpha3'],how='left')
len(df)
```

## 事業内容のカラムを追加する

```python
df_sentence = pd.read_csv('../out_csv/df_eval_sentences.csv')
df_sentence['description']=df_sentence['description']+df_sentence['background']
df=df.drop(['join_flg','Unnamed: 0'], axis=1)
df = pd.merge(df, df_sentence, on=['file'], how='left')
len(df)
```

## 成功失敗の2値ラベルを付与

```python
import numpy as np

LABEL_BINARY_COL = "eval_binary"      # 出力先（存在しなければ作る）
SOURCE_ORDINAL_COL = "total_eval"     # 1〜4等の元列名（なければ None）
SUCCESS_SET = {3, 4}                  # 成功とみなす元ラベル
FAIL_SET    = {1, 2}                  # 失敗とみなす元ラベル


# 二値ラベルを用意（eval_binary が無ければ total_eval 等から作成）
if LABEL_BINARY_COL not in df.columns:
    if SOURCE_ORDINAL_COL is None or SOURCE_ORDINAL_COL not in df.columns:
        raise ValueError(
            f"二値ラベル {LABEL_BINARY_COL} が無く、元列 {SOURCE_ORDINAL_COL} も見つかりません。"
            " LABEL_BINARY_COL を既存列にするか、SOURCE_ORDINAL_COL を正しく指定してください。"
        )
    # 数値化してから 3,4→1 / 1,2→0 / それ以外は NaN
    src = pd.to_numeric(df[SOURCE_ORDINAL_COL], errors="coerce")
    df[LABEL_BINARY_COL] = np.where(
        src.isin(SUCCESS_SET), 1,
        np.where(src.isin(FAIL_SET), 0, np.nan)
    )
```

```python
df.to_csv('../df_check_ml.csv', index=False)
```
